```python
import pandas as pd
import numpy as np

data = pd.read_csv("https://raw.githubusercontent.com/Salina-Huang/Programming-for-AI-Sleep-Health-and-Lifestyle-Model/refs/heads/main/Sleep_health_and_lifestyle_dataset.csv")
print(data.isnull().sum(), "\n")
print("*"*50)
print("The sum of duplicated is:", data.duplicated().sum(), "\n")
print("*"*50)
data.dropna()
print(data.isnull().sum())

data_features = data.copy()
data_split = data_features["Blood Pressure"].str.split("/", expand=True)
data_features["systolic"] = pd.to_numeric(data_split[0], errors="coerce")
data_features["diastolic"] = pd.to_numeric(data_split[1], errors="coerce")
data_features = data_features.drop('Blood Pressure', axis=1)


conditions = (
    # Normal
    (data_features["systolic"] < 120) & (data_features["diastolic"] <= 80),
    # Elevated
    (data_features["systolic"] >= 120) & (data_features["systolic"] <= 129) & (data_features["diastolic"] <= 80),
    # High
    (data_features["systolic"] > 129) & (data_features["systolic"] < 180) | (data_features["diastolic"] > 80) & (data_features["diastolic"] <= 120)
)

choices = ["Normal", "Elevated", "High"]
data_features["bp_category"] = np.select(conditions, choices, default="Hypertensive")
data_features = data_features.drop('systolic', axis=1)
data_features = data_features.drop('diastolic', axis=1)
print(data_features)
```
